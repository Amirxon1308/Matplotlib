## Ko'p grafiklarni tasvirlash

```python
#subplot() funksiyasi yordamida biz bitta figurada turli grafiklarni chizishimiz mumkin.
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1) # grafikda 2 ta qator, 1 ta ustun va ustundagi 1 rasm
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2) # grafikda 2 ta qator, 1 ta ustun va ustundagi 2 rasm
plt.plot(x,y)

plt.show()
```

### Title va Super Title 

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Sotish")  #subtitle

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 5, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.title("Foyda") #subtitle

plt.suptitle("Mening Do'konim")

plt.show()
```
